# Penugasan 04 - Modul Buatan Sendiri

```python
%%writefile konversi.py
def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    return (f - 32) * 5/9
```

```python
from konversi import celcius_ke_fahrenheit, fahrenheit_ke_celcius

print(celcius_ke_fahrenheit(30))
print(fahrenheit_ke_celcius(86))
```

# Penugasan 06 - Studi Kasus Mini

```python
%%writefile laporan.py
def tampilkan_laporan(total_soal, jumlah_benar):
    skor = int(jumlah_benar / total_soal * 100)
    print("Ringkasan Kuis")
    print("Total soal:", total_soal)
    print("Jawaban benar:", jumlah_benar)
    print("Skor:", skor)
```

```python
import random
from laporan import tampilkan_laporan

total_soal = 5
jumlah_benar = 0

for i in range(total_soal):
    a = random.randint(1, 10)
    b = random.randint(1, 10)
    print("Soal", i + 1, ":", a, "+", b, "=")
    jawaban = int(input("Jawab: "))
    if jawaban == a + b:
        jumlah_benar += 1

tampilkan_laporan(total_soal, jumlah_benar)
```
