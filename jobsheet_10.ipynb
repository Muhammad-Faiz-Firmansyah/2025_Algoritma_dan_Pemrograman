# Praktikum 01 – Bubble Sort – Gelembung yang Naik ke Atas

```python
def bubble_sort(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j] > data[j + 1]:
        data[j], data[j + 1] = data[j + 1],data [j]
  return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

# Praktikum 02 – Selection Sort

```python
def selection_sort(data):
  n = len(data)
  for i in range(n):
    min_index = i
    for j in range(i + 1, n):
      if data[j] < data[min_index]:
        min_index = j
    data[i], data[min_index] = data[min_index], data[i]
  return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

# Praktikum 03 – Insertion Sort

```python
def insertion_sort(data):
  n = len(data)
  for i in range(1,n):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

# Praktikum 04 – Studi Kasus: Mengurutkan Nilai Mahasiswa

```python
def bubble_sort_nilai(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j][1] < data[j + 1][1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

mahasiswa = [("Andi",80), ("Budi", 70), ("cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

# Praktikum 05 – Perbandingan Waktu Eksekusi Sorting

```python
import random, time
if 'bubble_sort' not in globals():
  def bubble_sort(data):
    n = len(data)
    for i in range(n):
      for j in range(0, n - i - 1):
        if data[j] > data[j + 1]:
          data[j], data[j + 1] = data[j + 1], data[j]
    return data

if 'selection_sort' not in globals():
  def selection_sort(data):
    n = len(data)
    for i in range(n):
      min_index = i
      for j in range(i + 1, n):
        if data[j] < data[min_index]:
          min_index = j
      data[i], data[min_index] = data[min_index], data[i]
    return data

if 'insertion_sort' not in globals():
  def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
      key = data[i]
      j = i - 1
      while j >= 0 and data[j] > key:
        data[j + 1] = data[j]
        j -= 1
      data[j + 1] = key
    return data

data = [random.randint(1, 1000) for _ in range(500)]

for func in [bubble_sort, selection_sort, insertion_sort]:
  salinan = data.copy()
  start = time.time()
  func(salinan)
  end = time.time()
  print(f"{func.__name__} membutuhkan waktu {end - start:.5f} detik")
```

# Praktikum 06 – Sorting Data Barang (Nama & Harga)

```python
def bubble_sort_key(data, key_index=0, reverse=False):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if reverse:
        if data[j][key_index] < data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
      else:
        if data[j][key_index] > data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
  return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

# Penugasan 05 - Analisis Kompleksitas

```python
def bubble_sort_count(data):
    n = len(data)
    count = 0
    for i in range(n):
        for j in range(0, n-i-1):
            count += 1
            if data[j] > data[j+1]:
                data[j], data[j+1] = data[j+1], data[j]
    return count

def selection_sort_count(data):
    n = len(data)
    count = 0
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            count += 1
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return count

def insertion_sort_count(data):
    n = len(data)
    count = 0
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            count += 1
            data[j + 1] = data[j]
            j -= 1
        if j >= 0:
            count += 1
        data[j + 1] = key
    return count

import random

sizes = [10, 50, 100]

print("ANALISIS KOMPLEKSITAS: JUMLAH PERBANDINGAN")
print("=" * 60)

for size in sizes:
    print(f"\nUkuran data: {size}")
    print("-" * 30)

    # Generate data acak
    data_bubble = [random.randint(1, 1000) for _ in range(size)]
    data_selection = data_bubble.copy()
    data_insertion = data_bubble.copy()

    # Hitung perbandingan
    count_bubble = bubble_sort_count(data_bubble.copy())
    count_selection = selection_sort_count(data_selection.copy())
    count_insertion = insertion_sort_count(data_insertion.copy())

    print(f"Bubble Sort: {count_bubble} perbandingan")
    print(f"Selection Sort: {count_selection} perbandingan")
    print(f"Insertion Sort: {count_insertion} perbandingan")
```

# Penugasan 06 - Proyek Mini

```python
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(n):
        for j in range(0, n-i-1):
            if reverse:
                if data[j][key_index] < data[j+1][key_index]:
                    data[j], data[j+1] = data[j+1], data[j]
            else:
                if data[j][key_index] > data[j+1][key_index]:
                    data[j], data[j+1] = data[j+1], data[j]
    return data

# Input daftar barang
barang = []
print("Masukkan daftar barang (ketik 'selesai' untuk berhenti):")

while True:
    nama = input("Nama barang (atau 'selesai'): ").strip()
    if nama.lower() == 'selesai':
        break

    try:
        harga = int(input("Harga barang: "))
        barang.append((nama, harga))
    except ValueError:
        print("Harga harus angka! Coba lagi.")

# Tampilkan daftar barang awal
print("\nDaftar barang awal:")
for i, (nama, harga) in enumerate(barang, 1):
    print(f"{i}. {nama} - Rp{harga}")

# Menu pengurutan
while True:
    print("\n=== MENU PENGURUTAN ===")
    print("1. Urutkan berdasarkan nama (A-Z)")
    print("2. Urutkan berdasarkan nama (Z-A)")
    print("3. Urutkan berdasarkan harga (terendah-tertinggi)")
    print("4. Urutkan berdasarkan harga (tertinggi-terendah)")
    print("5. Keluar")

    pilihan = input("Pilih opsi (1-5): ").strip()

    if pilihan == '1':
        hasil = bubble_sort_key(barang.copy(), 0, False)
        print("\nHasil pengurutan berdasarkan nama (A-Z):")

    elif pilihan == '2':
        hasil = bubble_sort_key(barang.copy(), 0, True)
        print("\nHasil pengurutan berdasarkan nama (Z-A):")

    elif pilihan == '3':
        hasil = bubble_sort_key(barang.copy(), 1, False)
        print("\nHasil pengurutan berdasarkan harga (terendah-tertinggi):")

    elif pilihan == '4':
        hasil = bubble_sort_key(barang.copy(), 1, True)
        print("\nHasil pengurutan berdasarkan harga (tertinggi-terendah):")

    elif pilihan == '5':
        print("Terima kasih!")
        break

    else:
        print("Pilihan tidak valid!")
        continue

    # Tampilkan hasil
    for i, (nama, harga) in enumerate(hasil, 1):
        print(f"{i}. {nama} - Rp{harga}")
```
