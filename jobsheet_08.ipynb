# Praktikum 04 - Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
    return sum(args)

def biodata(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

# Praktikum 07 - Fungsi Decorator

```python
def decorator_sapa(fungsi):
    def wrapper():
        print("Sebelum fungsi dijalankan...")
        fungsi()
        print("Setelah fungsi dijalankan...")
    return wrapper

@decorator_sapa
def sapa():
    print("Halo, Mahasiswa Polines!")

sapa()
```

# Praktikum 08 - Studi Kasus Menghitung Rata-rata Nilai Mahasiswa

# penugasan 08 - Decorator

```python
def logger(fungsi):
    def wrapper():
        print("Memulai fungsi...")
        fungsi()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo! Selamat datang di Politeknik Negeri Semarang!")

halo()
```

# penugasan 09 - Studi Kasus Program Nilai Mahasiswa

```python
def input_nilai():
    tugas = float(input("Masukkan nilai tugas: "))
    ujian = float(input("Masukkan nilai ujian: "))
    return [tugas, ujian]

def hitung_rata(daftar_nilai):
    return sum(daftar_nilai) / len(daftar_nilai)

def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak lulus"

nilai_akhir = lambda tugas, ujian: 0.4 * tugas + 0.6 * ujian

nilai = input_nilai()
tugas, ujian = nilai[0], nilai[1]

rata = hitung_rata(nilai)
akhir = nilai_akhir(tugas, ujian)
status = status_kelulusan(akhir)

print("Nilai rata-rata:", rata)
print("Nilai akhir (40% tugas, 60% ujian):", akhir)
print("Status kelulusan:", status)
```
