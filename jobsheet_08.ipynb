# Praktikum 04 - Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
    return sum(args)

def biodata(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

# Praktikum 05 - Scope Variabel

```python
x = 10  # global

def ubah():
    x = 5  # lokal
    print("Nilai lokal:", x)

ubah()
print("Nilai global:", x)
```

# Praktikum 07 - Fungsi Decorator

```python
def decorator_sapa(fungsi):
    def wrapper():
        print("Sebelum fungsi dijalankan...")
        fungsi()
        print("Setelah fungsi dijalankan...")
    return wrapper

@decorator_sapa
def sapa():
    print("Halo, Mahasiswa Polines!")

sapa()
```

# penugasan 05 - Fungsi dengan **kwargs

```python
def info_barang(**data):
    for key, value in data.items():
        print(f"{key}: {value}")

print("=== Info Barang 1 ===")
info_barang(Nama="Buku", Harga=15000, Stok=30)

print("\n=== Info Barang 2 ===")
info_barang(Nama="Pulpen", Harga=3000, Stok=100, Warna="Biru")
```

# penugasan 06 - Scope Variabel

```python
x = 100

def ubah():
    x = 50
    print("Nilai x di dalam fungsi (lokal):", x)

print("Nilai x sebelum pemanggilan fungsi (global):", x)
ubah()
print("Nilai x setelah pemanggilan fungsi (global):", x)

print("\n=== PENJELASAN ===")
print("Variabel x di dalam fungsi adalah variabel LOKAL yang hanya berlaku di dalam fungsi.")
print("Variabel x di luar fungsi adalah variabel GLOBAL yang dapat diakses dari mana saja.")
print("Meskipun namanya sama, keduanya adalah variabel yang berbeda dan tidak saling mempengaruhi.")
```

# penugasan 07 - Fungsi Lambda

```python
import math

kuadrat = lambda x: x ** 2

print("=== Lambda Kuadrat ===")
print("Kuadrat dari 5:", kuadrat(5))
print("Kuadrat dari 8:", kuadrat(8))
print("Kuadrat dari 12:", kuadrat(12))

keliling_lingkaran = lambda r: 2 * math.pi * r

print("\n=== Lambda Keliling Lingkaran ===")
print(f"Keliling lingkaran dengan r=7: {keliling_lingkaran(7):.2f}")
print(f"Keliling lingkaran dengan r=10: {keliling_lingkaran(10):.2f}")
print(f"Keliling lingkaran dengan r=14: {keliling_lingkaran(14):.2f}")
```

# penugasan 08 - Decorator

```python
def logger(fungsi):
    def wrapper():
        print("Memulai fungsi...")
        fungsi()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo! Selamat datang di Politeknik Negeri Semarang!")

halo()
```

# penugasan 09 - Studi Kasus Program Nilai Mahasiswa

```python
def input_nilai():
    nilai = []
    jumlah = int(input("Berapa jumlah nilai yang akan diinput? "))
    for i in range(jumlah):
        n = float(input(f"Masukkan nilai ke-{i+1}: "))
        nilai.append(n)
    return nilai

def hitung_rata(data):
    return sum(data) / len(data)

def status_kelulusan(rata):
    if rata >= 60:
        return "LULUS"
    else:
        return "TIDAK LULUS"

nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

print("="*50)
print("PROGRAM PERHITUNGAN NILAI MAHASISWA")
print("="*50)

print("\n--- Contoh 1: Mahasiswa Faiz ---")
nilai_faiz = [80, 85, 90, 75, 88]
rata_faiz = hitung_rata(nilai_faiz)
status_faiz = status_kelulusan(rata_faiz)

print(f"Nilai-nilai: {nilai_faiz}")
print(f"Rata-rata: {rata_faiz:.2f}")
print(f"Status: {status_faiz}")

print("\n--- Contoh 2: Mahasiswa Budi ---")
tugas_budi = 75
ujian_budi = 85
akhir_budi = nilai_akhir(tugas_budi, ujian_budi)
status_budi = status_kelulusan(akhir_budi)

print(f"Nilai Tugas: {tugas_budi}")
print(f"Nilai Ujian: {ujian_budi}")
print(f"Nilai Akhir (40% tugas + 60% ujian): {akhir_budi:.2f}")
print(f"Status: {status_budi}")

print("\n--- Contoh 3: Mahasiswa Citra ---")
nilai_citra = [50, 55, 45, 60, 48]
rata_citra = hitung_rata(nilai_citra)
status_citra = status_kelulusan(rata_citra)

print(f"Nilai-nilai: {nilai_citra}")
print(f"Rata-rata: {rata_citra:.2f}")
print(f"Status: {status_citra}")

print("\n" + "="*50)
```
